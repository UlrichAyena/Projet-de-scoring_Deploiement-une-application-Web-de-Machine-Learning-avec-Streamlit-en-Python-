 #### Explication du projet
 Qui est le plus susceptible d'avoir un compte bancaire ? L'inclusion financière reste l'un des principaux obstacles au développement économique et humain en Afrique. Par exemple, au Kenya, au Rwanda, en Tanzanie et en Ouganda, seuls 9,1 millions d'adultes (soit 13,9 % de la population adulte) ont accès à un compte bancaire commercial ou l'utilisent. Traditionnellement, l'accès à un compte bancaire est considéré comme un indicateur d'inclusion financière. Malgré la prolifération de l'argent mobile en Afrique, et la croissance des technologies innovantes. Les banques continuent de jouer un rôle essentiel pour faciliter l'accès aux services financiers. L'accès à la banque permet aux ménages d'épargner et de faciliter les paiements tout en aidant les entreprises à améliorer leur solvabilité et leur accès à d'autres services financiers. Par conséquent, l'accès aux comptes bancaires est un facteur essentiel de la croissance économique à long terme.
 
 #### Objetif du projet
 Il s’agit de développer une solution de scoring permettant d’apporter une aide à la décision pour les chargés de relation client. En effet, basée sur des algorithmes de Machine Learning, la solution calcule la probabilité qu’un client ait le droit d’ouvrir un compte bancaire ou non.

```python
from sklearn.preprocessing import LabelEncoder

dataset5 = dataset.drop(["uniqueid"], axis=1).copy()

# Création d'un objet LabelEncoder
label_encoder = LabelEncoder()
dataframe = dataset5

# Initialisation du compteur pour suivre le nombre de colonnes encodées
encoded_columns_count = 0

# Liste pour stocker les noms des variables encodées
encoded_columns = []

# Parcours de toutes les colonnes du DataFrame
for column_name in dataframe:
    # Vérification si le type de données de la colonne est 'object' (c.-à-d. une caractéristique catégorielle)
    if dataframe[column_name].dtype == 'object':
        # Vérification s'il y a 2 catégories uniques ou moins dans la colonne
        if len(dataframe[column_name].unique()) <= 2:
            # Entraînement de l'encodeur sur les données d'entraînement de cette colonne et transformation de la colonne
            label_encoder.fit(dataframe[column_name])
            dataframe[column_name] = label_encoder.transform(dataframe[column_name])
            # Incrément du compteur de colonnes encodées
            encoded_columns_count += 1
            # Ajout du nom de la colonne encodée à la liste
            encoded_columns.append(column_name)

# Affichage du nombre total de colonnes encodées
print('%d colonnes ont été encodées.' % encoded_columns_count)

# Affichage des noms des colonnes encodées
print('Colonnes encodées :', encoded_columns)
```

###  4. Modélisation et Prediction

```python
## Séparation de la varible à expliquer (target) et les variables explicatives(features)
y= df[TARGET]
X = df.drop(TARGET, axis=1)
X.head()
```

```python
## Division en données d'apprentissage et données test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=123)
```

```python
print(" Taille des X_train:",X_train.shape)
print(" Taille des X_train:",X_test.shape)
print(" Taille des X_train:",y_train.shape)
print(" Taille des X_train:",y_test.shape)
```

####  Mise en place des Modèles par défaut  
1 - Régression logistique

2 - Forêt aléatoire

3 - XGBClassifier

```python
##Normalisation des  variables continues
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train[VARIABLE_CONTINUES] = scaler.fit_transform(X_train[VARIABLE_CONTINUES])
X_test[VARIABLE_CONTINUES] = scaler.transform(X_test[VARIABLE_CONTINUES])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score,accuracy_score, f1_score

# Créer des instances des modèles
 
    #Régression logistique 
log_reg = LogisticRegression()
   # Forêt aléatoire
random_forest = RandomForestClassifier()
  # XGBClassifier
xgboost = XGBClassifier()

# Entraîner les modèles
log_reg.fit(X_train, y_train)
random_forest.fit(X_train, y_train)
xgboost.fit(X_train, y_train)

# Prédire les probabilités de classe 1 (avoir un compte bancaire)
log_reg_probs = log_reg.predict_proba(X_test)[:, 1]
random_forest_probs = random_forest.predict_proba(X_test)[:, 1]
xgboost_probs = xgboost.predict_proba(X_test)[:, 1]

# Calculer l'AUC pour chaque modèle
log_reg_auc = roc_auc_score(y_test, log_reg_probs)
random_forest_auc = roc_auc_score(y_test, random_forest_probs)
xgboost_auc = roc_auc_score(y_test, xgboost_probs)

# Créer un DataFrame pour stocker les résultats
results = pd.DataFrame({
    'Modèle': ['Régression Logistique', 'Forêt Aléatoire', 'XGBoost'],
    'AUC': [log_reg_auc, random_forest_auc, xgboost_auc]
})

# Afficher le DataFrame
print(results)
```

```python
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# Définir le nombre de folds pour la validation croisée
kfold = StratifiedKFold(n_splits=5, random_state=100, shuffle=True)

# Fonction pour la validation croisée
def cross_validation(x, y, model):
    result = cross_val_score(model, x, y, cv=kfold, scoring="roc_auc", n_jobs=-1)
    print("Score: %f" % result.mean())

# Fonction pour l'ajustement des hyperparamètres avec recherche aléatoire
def RndSrch_Tune(model, X, y, params):
    clf = RandomizedSearchCV(model, params, scoring='roc_auc', cv=kfold, n_jobs=-1, random_state=100)
    clf.fit(X, y)
    print("best score is :", clf.best_score_)
    print("best estimator is :", clf.best_estimator_)
    print("best Params is :", clf.best_params_)
    return (clf.best_estimator_)

# Modèles à utiliser
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier()
}

# Paramètres pour la recherche aléatoire

# Régression logistique
param_logistic = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l2']
}

# Forêt aléatoire
param_rf = {
    'n_estimators': [50, 100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# XGBClassifier
param_xgb = {
    'learning_rate': [0.01, 0.05, 0.1, 0.3],
    'max_depth': [3, 4, 5, 6, 8, 10],
    'n_estimators': [50, 100, 200, 300]
}

# Dictionnaire pour stocker les meilleurs modèles
best_models = {}

# Boucle sur les modèles
for model_name, model in models.items():
    print(f"Evaluating {model_name}:")
    best_model = RndSrch_Tune(model, X_train, y_train, param_logistic if model_name == "Logistic Regression"
                              else param_rf if model_name == "Random Forest" else param_xgb)
    cross_validation(X_train, y_train, best_model)
    best_models[model_name] = best_model

# Évaluation des performances des meilleurs modèles sur l'ensemble de test
print("\nEvaluation on test set:")
# Créer un DataFrame pour stocker les résultats
results_df = pd.DataFrame(columns=['Model', 'AUC Score'])
for model_name, model in best_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict_proba(X_test)[:, 1]
    auc_score = roc_auc_score(y_test, y_pred)
    # Ajouter les résultats au DataFrame
    results_df = results_df.append({'Model': model_name, 'AUC Score': auc_score}, ignore_index=True)
# Afficher le DataFrame
print(results_df)
```

```python
# Enregistrement chaque modèle avec les meilleurs paramètres
import joblib
for model_name, model in best_models.items():
    filename = f"{model_name}_best_model.joblib"
    joblib.dump(model, filename)
    print(f"{model_name} saved as {filename}")
```

- Tracé de la courbe ROC

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Créer une liste des meilleurs modèles
classifiers = [best_models[model_name] for model_name in best_models]

# Créer un DataFrame pour stocker les résultats
result_table = pd.DataFrame(columns=['classifiers', 'fpr', 'tpr', 'auc'])

# Boucle sur les modèles pour calculer les FPR, TPR et AUC
for model_name, model in best_models.items():
    # Prédire les probabilités de classe 1 sur l'ensemble de données de test
    y_probs = model.predict_proba(X_test)[:, 1]
    # Calculer le taux de faux positifs (FPR), le taux de vrais positifs (TPR) et l'AUC
    fpr, tpr, _ = roc_curve(y_test, y_probs)
    auc_score = auc(fpr, tpr)
    # Ajouter les résultats au DataFrame
    result_table = result_table.append({'classifiers': model_name,
                                        'fpr': fpr,
                                        'tpr': tpr,
                                        'auc': auc_score}, ignore_index=True)

# Tracer la courbe ROC pour chaque modèle
plt.figure(figsize=(10, 8))
for i in range(result_table.shape[0]):
    plt.plot(result_table.loc[i, 'fpr'], result_table.loc[i, 'tpr'],
             label=f"{result_table.loc[i, 'classifiers']} (AUC = {result_table.loc[i, 'auc']:.2f})")

# Tracer la ligne en pointillés diagonale (aucun skill)
plt.plot([0, 1], [0, 1], linestyle='--', color='orange')

# Ajouter des labels et un titre
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.title('Courbe ROC pour les modèles')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

### 5. Choix du bon modèle
